# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Изучение данных из файла

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/datasets/real_estate_data.csv',sep='\t')
df.info()
df.columns
```

### Вывод

Некоторые колонки имеют неверный тип данных и имеют пропущенные значения.

## Предобработка данных

Удалим явные дубликаты

```python
df = df.drop_duplicates()
```

Смотрим % по пропущеным значениям

```python
total_nan = df.isna().sum() * 100 / len(df) 
display(total_nan.sort_values()) # Процент пропусков по коллонкам
```

### rooms

```python
df['rooms'].unique()
```

У нас имеется недвижимость где 0 комнат. Возможно это студии.

```python
zero_rooms = df.loc[df['rooms']==0]
zero_rooms.groupby('studio')['rooms'].count()
print(zero_rooms[zero_rooms['studio']==False]['rooms'].count().sum() * 100 / len(df))
```

Всего 59 записей где комнат 0 и при этом не студия. что меньше 1% от общих данных. По этому для чистоты анализа удаляем такие записи.

```python
df.drop(df.loc[(df.rooms ==0)& (df.studio == False)].index,inplace = True)
df.loc[(df['rooms']==0)&(zero_rooms['studio']==False)]
```

0 строк, успешно удалено. Теперь мы точно знаем что там где 0 комнат у нас студия.

### floors_total

Колонка floors_total имеет тип данных float64 что не верно. В этой колонке числа целые. Так как строк с пропущенным значением меньше 1% то имеет смысл их просто удалить.

```python
df = df.dropna(subset=['floors_total'])
df.info()
```

Строки удалены, строк осталось 23506

### floor

```python
sorted(df['floor'].unique())
```

### is_apartment

```python
df['is_apartment'].unique()
```

Колонка булевого типа, но имеет строки с NaN

```python
df.loc[(df['is_apartment'].isna())]
```

Апартаменты  это автономная единица жилья (тип жилой недвижимости ), которая занимает часть здания, как правило, на одном этаже . Так как 20866 строк не указали значение, то можно по умолчанию ставить False

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

```python
df['is_apartment'].unique()
df.info()
```

### kitchen_area

В этой ячейке решением по заполнению пропусков я вижу так:
необходимо сгруппировать данные по total_area и для каждой группы вычислить медиану по kitchen_area. Этими данными заполнить пропуски

```python
df['total_area_group'] = pd.cut(df['total_area'],bins=5)
df['kitchen_area'] = df['kitchen_area'].fillna(df.groupby('total_area_group')['kitchen_area'].transform('median'))
```

```python
df.info()
```

Все заполнено.

### living_area

Аналогично поступим и для living_area

```python
df['living_area'] = df['living_area'].fillna(df.groupby('total_area_group')['living_area'].transform('median'))
```

```python
df.info()
```

```python
sorted(df['balcony'].unique())
```

```python
df['balcony'] = df['balcony'].fillna(0)
df['balcony'] = df['balcony'].astype('int')
```

```python
df.info()
```

тип стал int64. NaN нету все ок.

## Расчёты и добавление результатов в таблицу

#### цена квадратного метра

```python
df['one_quad_meter'] = df['last_price']/df['total_area']
```

Посчитано и добавлено в таблицу: цена квадратного метра

#### день недели, месяц и год публикации объявления

```python
df['Date_first_day_exposition'] = df['first_day_exposition'].dt.strftime('%d/%m/%Y')
```

Посчитано и добавлено в таблицу: день недели, месяц и год публикации объявления

#### этаж квартиры, варианты — первый, последний, другой

```python
df.loc[df['floor']==df['floors_total'], 'Floor_category'] = 'последний'
df.loc[df['floor']==df['floors_total']/df['floors_total'], 'Floor_category'] = 'первый'
df['Floor_category'] = df['Floor_category'].fillna('другой')
```

Посчитано и добавлено в таблицу: этаж квартиры; варианты — первый, последний, другой

#### соотношение жилой и общей площади

```python
df['living_area_ratio'] = df['living_area']/df['total_area']*100
```

Посчитано и добавлено в таблицу: соотношение жилой и общей площади

#### отношение площади кухни к общей площади

```python
df['kitchen_area_ratio'] = df['kitchen_area']/df['total_area']*100
```

Посчитано и добавлено в таблицу: отношение площади кухни к общей

####  удаленность до центра в киллометрах

```python
df['cityCenters_nearest_km'] = df['cityCenters_nearest'] / 1000
df['cityCenters_nearest_km'] = df['cityCenters_nearest_km'].round()
```

Колонка удаленности до центра в киллометрах добавлена.

#### цена одного километра от центра

```python
df['city_km_price'] = df['last_price'] / df['cityCenters_nearest_km']
df['city_km_price'] = df['city_km_price'].round()
```

Цена одного километра от центра добавлена
